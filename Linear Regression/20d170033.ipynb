```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
file = pd.read_csv("/content/drive/MyDrive/Semesters/Semester 3 2021 Autumn/DS 203/Assignments/Linear Regression/Real_estate_valuation_data_set.csv")
```

```python
display(file)
```

## Training model using 80:20 train/test split

```python
#Preliminaries
file['X7 transaction year'] = np.array(file['X1 transaction date']).astype(int) #making seperate column for year
file['X8 transaction month'] = np.rint(((np.array(file['X1 transaction date']) - np.array(file['X1 transaction date']).astype(int))*12)).astype(int) #making seperate column for month
#subtracting the integer part and multiplying fractional part by 12 and taking closest integer value
del file['X1 transaction date']
# removing the X1 transaction date column because of remaining two columns

xlabels = ['X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude',	'X6 longitude','X7 transaction year','X8 transaction month']
ylabels = ['Y house price of unit area']
train = file.sample(frac = 0.8)
#randomly sampling 80% of the dataset
test = file.drop(train.index)
#sampling the remaining 20% that were not chosen 
display(train)
display(test)
```

```python
linear_model = LinearRegression()
#creating object
```

## Coefficients of Trained Model

```python
linear_model.fit(train[xlabels],train[ylabels])
#fitting with training model
#printing linear parameters
for i in range(len(ylabels)):
  print("Intercept for",ylabels[i],":", linear_model.intercept_[i])

for i in range(len(xlabels)):
  print("Slope for ",xlabels[i], linear_model.coef_[0][i])
```

## Predictions for Test Data


```python
 #predicting using the model
 prediction = linear_model.predict(test[xlabels])
 no_entries = len(prediction)
 test["Y1"] = prediction
```

```python
def display_predictions():
  #function to display the predicted values
  l = list(test['Y1'])
  j = list(test['Y house price of unit area'])
  print("Actual","|","Predicted")
  for i in range(len(l)):
    print(j[i],"|",l[i])

display_predictions()
```

<h2>Mean Square Error and $R^2$ Calculation

```python
y = np.array(test[['Y house price of unit area']]) #actual values
p_y = np.array(test[['Y1']]) #predicted values

mse_y = (np.square(y- p_y)).mean() #mean square error

print("Mean Square Error of Model:",mse_y)
```

```python
from sklearn.metrics import r2_score
#using an inbuilt function for the r^2 score which is 1 - RSS/TSS
r2 = r2_score(y, p_y)
print("R^2 score for this Linear Regression:",r2)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
lambda_values = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5]
#An array with lambda values
train_test_ratios = [0.8, 0.6, 0.7, 0.9]
#An array with train:test ratios
for lambda_val in lambda_values:
  for ratio in train_test_ratios:
    #iterating through
    print('\n')
    print("For lambda =",lambda_val,"and train:test ratio",ratio)
    xlabels = ['X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude',	'X6 longitude','X7 transaction year','X8 transaction month']
    ylabels = ['Y house price of unit area']
    line_model = Ridge(alpha = lambda_val) #creating ridge object with lambda(here alpha) value equal to lambda_val
    #sampling data as before
    train = file.sample(frac = ratio) 
    test = file.drop(train.index)
    line_model.fit(train[xlabels],train[ylabels]) #training model
    prediction = linear_model.predict(test[xlabels]) #predicting using model
    no_entries = len(prediction)
    test["Y1"] = prediction

    #printing relevant outputs
    y = np.array(test[['Y house price of unit area']])
    p_y = np.array(test[['Y1']]) 
    mse_y = (np.square(y- p_y)).mean()
    print("Mean Square Error of Model:",mse_y)
    from sklearn.metrics import r2_score
    r2 = r2_score(y, p_y)
    print("R^2 score for this Ridge Regression:",r2)

```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
lambda_values = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5]
#An array with lambda values
train_test_ratios = [0.8, 0.6, 0.7, 0.9]
#An array with train:test ratios
for lambda_val in lambda_values:
  for ratio in train_test_ratios:
    print('\n')
    print("For lambda =",lambda_val,"and train:test ratio",ratio)
    xlabels = ['X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude',	'X6 longitude','X7 transaction year','X8 transaction month']
    ylabels = ['Y house price of unit area']
    line_model = Lasso(alpha = lambda_val) #creating lasso object with lambda(here alpha) value equal to lambda_val
    train = file.sample(frac = ratio)
    test = file.drop(train.index)

    line_model.fit(train[xlabels],train[ylabels]) #training the model
    prediction = linear_model.predict(test[xlabels]) #predicting Y
    no_entries = len(prediction)
    test["Y1"] = prediction
    #printing various values
    y = np.array(test[['Y house price of unit area']])
    p_y = np.array(test[['Y1']]) 
    mse_y = (np.square(y- p_y)).mean()

    print("Mean Square Error of Model:",mse_y)

    from sklearn.metrics import r2_score
    r2 = r2_score(y, p_y)
    print("R^2 score for this Lasso Regression:",r2)
```

```python
# Preliminaries

# making a dictionary for column names as the given file doesnt have column class names
data_columns={0: 'Age of patient at the time of operation', 1: "Patient's year of operation", 2:"Number of positive auxillary nodes detected",3:"Survival status"}
file = pd.read_csv("/content/drive/MyDrive/Semesters/Semester 3 2021 Autumn/DS 203/Assignments/Linear Regression/haberman.csv", header=None)
file.rename(columns = data_columns, inplace=True) #renaming the first column to the dictionary made above
```

```python
display(file)
```

```python
xlabels = list(data_columns.values())[:-1] # taking all but last column name
ylabels = list(data_columns.values())[-1]  # taking the last column name 

train = file.sample(frac = 0.8) #sampling 80% of the data for training
test = file.drop(train.index) #taking rest for testing

display(train)
display(test)
```

```python
from sklearn.linear_model import LogisticRegression
linear_model = LogisticRegression()
linear_model.fit(train[xlabels],train[ylabels]) #training model

#Printing model properties
print("Intercept for",ylabels,":", linear_model.intercept_[0])

for i in range(len(xlabels)):
  print("Slope for ",xlabels[i], linear_model.coef_[0][i])

```

```python
prediction = linear_model.predict(test[xlabels])
test['Predicted Survival'] = prediction

display(test)
```

### Evaluating Classification Error

```python
def classification_error(y,y_pred):
  #defining this function which counts number of unequal outputs and divides by total number of outputs
  count = 0
  m = len(y)
  for i in range(m):
    if(y[i]!=y_pred[i]):
      count += 1
  return (count/m)

print("Classification Error for this Logistic Regression is",classification_error(list(test['Survival status']),list(test['Predicted Survival'])))
```

```python
from sklearn.linear_model import LogisticRegression
#varying over different train:test ratios and performing same operation as above for each
test_train = [0.6, 0.7, 0.9]
for ratio in test_train:
  xlabels = list(data_columns.values())[:-1]
  ylabels = list(data_columns.values())[-1]
  train = file.sample(frac = ratio)
  test = file.drop(train.index)
  linear_model = LogisticRegression()
  linear_model.fit(train[xlabels],train[ylabels])
  print("For train:test ratio",ratio)
  print("Intercept for",ylabels,":", linear_model.intercept_[0])
  for i in range(len(xlabels)):
    print("Slope for ",xlabels[i], linear_model.coef_[0][i])
  prediction = linear_model.predict(test[xlabels])
  test['Predicted Survival'] = prediction
  print("Classification Error for this Logistic Regression is",classification_error(list(test['Survival status']),list(test['Predicted Survival'])),'\n')
```
